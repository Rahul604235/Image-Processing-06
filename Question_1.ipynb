# Implementing 2D DFT using row-column decomposition.

```python
# Numpy, Pillow, cmath, math libraries were used to do the following program. 

# forces floating point division
from __future__ import division 
 
import numpy as np
from PIL import Image
import math
import cmath
```

# DFT Image1

```python
# fourier_image was opened using pillow library.
fourier=Image.open('fourier.png')
```

```python
#This is fourier_image
fourier
```

```python
#Converts image into matrix.
Fmn_matrix1=np.array(fourier)
```

```python
#Fmn_matrix
Fmn_matrix1
```

```python
#M,N siginifies rows and columns of the Fmn_matrix1 using shape function.
M,N=Fmn_matrix1.shape
```

```python
M,N
```

```python
#Zero matrix of M*N is created and assigned as Fml_matrix1.
Fml_matrix1=np.zeros((M,N),dtype=complex)
M,L=Fml_matrix1.shape

#Zero matrix of M*N is created and assigned as Fkl_matrix1
Fkl_matrix1=np.zeros((M,N),dtype=complex)
K,L=Fkl_matrix1.shape
```

```python
#Fkl_matrix
Fkl_matrix1
```

```python
#This is the code used to find out 2D dft using 1D dft inbuild function.
for m in range(M):
    row_array=Fmn_matrix1[m,:]
    arr1=np.fft.fft(row_array)
    Fml_matrix1[m,:]=arr1
    
for l in range(L):
    column_array=Fml_matrix1[:,l]
    arr2=np.fft.fft(column_array)
    Fkl_matrix1[:,l]=arr2


#Code to find FFT shift.
Fkl_matrix_final1=np.zeros((M,N),dtype=complex)
Kf,Lf=Fkl_matrix_final1.shape

Fkl_matrix_final1[0:int(M/2),0:int(N/2)]=Fkl_matrix1[int(M/2):int(M),int(N/2):int(N)]
Fkl_matrix_final1[0:int(M/2),int(N/2):int(N)]=Fkl_matrix1[int(M/2):int(M),0:int(N/2)]
Fkl_matrix_final1[int(M/2):int(M),0:int(N/2)]=Fkl_matrix1[0:int(M/2),int(N/2):int(N)]
Fkl_matrix_final1[int(M/2):int(M),int(N/2):int(N)]=Fkl_matrix1[0:int(M/2),0:int(N/2)]


#This is the code to find the magnitude spectrum and log transform was used.
mag1_abs=np.abs(Fkl_matrix_final1)
max_value1=mag1_abs.max()
c=255/np.log(1+max_value1)
mag1=c*(np.log(1+mag1_abs))

#This is the code to find the Phase spectrum and angles are in degrees.
angle1=np.angle(Fkl_matrix_final1)
angle1=np.degrees(angle1)
```

```python
#Converts the mag1 to the Final_Image and Final_image is displayed (magnitude spectrum).
Final_image = Image.fromarray(mag1)
Final_image.show()
```

```python
#Converts the mag1 to the Final_Image and Final_image is displayed (Phase spectrum).
Final_image = Image.fromarray(angle1)
Final_image.show()
```
